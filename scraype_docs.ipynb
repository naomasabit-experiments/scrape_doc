```python
import urllib.request,http
import json,os,time
from urllib.parse import urlparse, quote_plus,unquote
BODY = "https://www.googleapis.com/customsearch/v1?"
#set key & cx & file_type & search_word
KEY ="key=hogehoge"
CX = "&cx=hogehoge"
FILE_TYPE="&fileType=docx"
SEARCH_WORD="&q=" + urllib.parse.quote_plus("hogehoge",encoding="utf-8")
SORT = "&sort=date"
START = "&start="
```

```python
def is_docx(tgt_path):
        root,ext = os.path.splitext(tgt_path)
        return ext == ".docx"
```

```python
def get_doc_lists(start_idx=1,end_idx=10):
    doc_list = [] 
    for i in range(start_idx,end_idx):
        #build URL
        url = BODY + KEY + CX + FILE_TYPE + SEARCH_WORD + SORT + START + str(i)
        #get search result by GOOGLE CUSTOM SEARCH API
        response = urllib.request.urlopen(url)
        #wait for request
        time.sleep(5)
        #get search result by JSON 　format
        data = json.loads(response.read().decode('utf-8'))
        #append document url list for search result
        for doc in data["items"]:
            link = doc["link"]
            #append only docx extention
            if is_docx(link):
                doc_list.append(link)
    return doc_list
```

```python
def download_doc_list(doc_list):
    for tgt in doc_list:
        try:
            file_name = urllib.parse.unquote(tgt.rsplit('/', 1)[1].split('?')[0])
            print("start download..." + file_name)
            urllib.request.urlretrieve(tgt,file_name)
            print(file_name + " has downloaded.")
        except:
            print("error")
```
